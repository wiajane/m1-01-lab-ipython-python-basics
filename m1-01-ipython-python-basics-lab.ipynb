```python
values=["apple", 178, False]
alias= values
alias.append(4)

print("Values:", values)
print("Alias:", alias)
```

```python
values_copy=list(values)
values_copy.append(103)

print("Values:", values)
print("Copy of values:", values_copy)
```

```python
record= {"name" : "jane", "age" : 20}
record_alias=record
record_alias["city"]= "Baku"

record_copy=record.copy()
record_copy["grade"]= "A"

print("record:", record)
print("alias:", record_alias)
print("Copy of record:", record_copy)
```

Variables are just references that are tagged to objects in memory. 
When we assign values/records to aliases, we add a label to the same object. So whenever one is changed, the other one is changed too. 
But when we use copy methods, we create a new object and assign the values of the corresponding variables. So they are independent from the original ones. 

How that affect passing objects into functions:
When we assign a new value to the variable using = inside the function, it can break the link to the original object. The local variable is going to point to smth new and the original remains the same.

```python
dict=[
    {"user_id": 1, "event_type": "pull", "duration_seconds": 180},
    {"user_id": 2, "event_type": "push", "duration_seconds": 100},
    {"user_id": 3, "event_type": "fork", "duration_seconds": "Null"}, 
    {"user_id": 4, "event_type": "pull", "duration_seconds": 97},
    {"user_id": 5, "event_type": "pull", "duration_seconds": 56},
    {"user_id": 6, "event_type": "push", "duration_seconds": 104},
    {"user_id": 7, "event_type": "merge", "duration_seconds": 178},
    {"user_id": 8, "event_type": "push", "duration_seconds": -42},   
    {"user_id": 9, "event_type": "merge", "duration_seconds": 57},
    {"user_id": 10, "event_type": "fork", "duration_seconds": 215},
    {"user_id": 11, "event_type": "push", "duration_seconds": 175},
    {"user_id": 12, "event_type": "merge", "duration_seconds": 58}
]
cleaned_events = []

for e in dict:
    d = e["duration_seconds"]

    if type(d) == int or type(d) == float:
        if d >= 0:
            duration_minutes = d / 60
            e["duration_minutes"] = duration_minutes
            cleaned_events.append(e)

print("Valid record count:" , len(cleaned_events)) 
print("Keys for first record to check:", cleaned_events[0].keys())

print("For inspection:")
print(cleaned_events[0])
print(cleaned_events[-1])
```

TASK 5

```python
event_counts = {}
duration_sums = {}

for record in cleaned_events:
    etype = record["event_type"]
    dur = record["duration_minutes"]
    
    if etype not in event_counts:
        duration_sums[etype] = 0
        event_counts[etype] = 0
    
    duration_sums[etype] += dur
    event_counts[etype] += 1

    
average_duration = {}

for etype in duration_sums:
    average_duration[etype] = duration_sums[etype] / event_counts[etype]

   
unique_users = set()

for record in cleaned_events:
    unique_users.add(record["user_id"])

unique_user_count = len(unique_users)

total_count = sum(event_counts.values())


csv_lines = ["metric,key,value"]

for et in event_counts:
    csv_lines.append(f"count,{etype},{event_counts[etype]}")

for et in average_duration:
    csv_lines.append(f"average_duration,{etype},{average_duration[etype]}")

csv_lines.append(f"unique_users,all,{unique_user_count}")

print(csv_lines)
```
